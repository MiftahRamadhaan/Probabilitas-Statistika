```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
data_rama = pd.read_clipboard(sep=',')
data_rama.head(5)
```

```python
col_list=['Price']
numhouse = data_rama[data_rama.columns[data_rama.columns.isin(col_list)]] 
plt.figure(figsize=(10,5))
numhouse.boxplot(sym='r*', grid=False) 
plt.show()
```

```python
plt.figure(figsize=(15,5)) 
plt.subplot(121)
data_rama['Price'].plot.hist(bins=10, title='Price') 
plt.show()
```

```python
col_list=['Price', 'SqFt']
numhouse = data_rama[data_rama.columns[data_rama.columns.isin(col_list)]] 
numhouse.plot.scatter(x='SqFt', y='Price')
```

```python
col_list=['Price', 'Bedrooms']
numhouse = data_rama[data_rama.columns[data_rama.columns.isin(col_list)]] 
plt.figure(figsize=(15,5))
numhouse.boxplot(by='Bedrooms') 
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data_rama['Bedrooms'], data_rama['Price'])
plt.title('Scatter Plot Harga Rumah Berdasarkan Bedrooms')
plt.xlabel('Jumlah Bedrooms')
plt.ylabel('Harga Rumah')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(data_rama['Bathrooms'], data_rama['Price'], alpha=0.2)
plt.title('Scatter Plot Harga Rumah Berdasarkan Bathrooms')
plt.xlabel('Jumlah Bathdrooms')
plt.ylabel('Harga Rumah')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(5, 5))
sns.boxplot(x='Bathrooms', y='Price', data=data_rama)
plt.title('Grouped Boxplot Harga Rumah Berdasarkan Bathrooms')
plt.xlabel('Jumlah Bathrooms')
plt.ylabel('Harga Rumah')
plt.xticks(rotation=90) 
plt.grid(True)
plt.show()
```
